# 韓流分析
### 有鑑於"韓流"席捲各大場所(含台大學餐)
### 讓我們分析一下韓流的起源地

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.font_manager import FontProperties
```

## 讀取中選會資料
[中央選舉委員會] (http://db.cec.gov.tw/histQuery.jsp?voteCode=20181101A1B1&qryType=ctks&prvCode=64&cityCode=000&candNo=1)

```python
df = pd.read_html("http://db.cec.gov.tw/histQuery.jsp?voteCode=20181101A1B1&qryType=ctks&prvCode=64&cityCode=000&candNo=1")
```

```python
df_all = df[0]
df_all
```

```python
colnames = df_all.iloc[0].values
```

```python
df_all = df_all[1:]
df_all.columns = colnames
```

## 去掉無用的姓名及號次

```python
df_all = df_all.drop(['姓名', '號次'], axis = 1)
```

```python
df_all
```

## 物件轉型

```python
regions = df_all.iloc[1:,0].values
votes = df_all.iloc[1:, 1].values.astype('int')
rating = df_all.iloc[1:, 2].values
```

## 切割贅字

```python
for i in range(regions.shape[0]):
    regions[i] = regions[i].split('高雄市')[1]
    regions[i] = regions[i].split('區')[0]
    rating[i] = rating[i].split('%')[0]
```

```python
rating = rating.astype('float')
```

```python
regions, votes, rating
```

```python
who_are_top5 = np.where(votes > 59000)
others = np.where(votes < 59000)
```

## 畫個柱狀圖看看

```python
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams.update({'font.size': 12})
bar = plt.bar(regions, votes, width = 0.5, color = 'b')
for i in range(5):
    bar[who_are_top5[0][i]].set_color('r')
fig = plt.gcf()
plt.title("高雄各區韓粉人數(前五名上紅色)", fontsize = 30)
plt.xlabel("地區(區)", fontsize = 20)
plt.ylabel("韓粉(人)", fontsize = 20)
locs, labels = plt.xticks()
locs=[i for i in range(0,38)]
labs=[i for i in range(0,38)]
for i, v in enumerate(votes):
    plt.text(locs[i] - 0.5, v, str(v), fontsize = 12)
fig.set_size_inches(20, 10)

# plt.pie(rating, labels = regions, autopct = '%1.0f%%')
# figPie.set_size_inches(18.5, 10.5)
```

## 列出前三最高信仰的地區

```python
df_top1 = df_all.sort_values(by = ['得票率'], ascending = False)[0:1]
df_top2 = df_all.sort_values(by = ['得票率'], ascending = False)[1:2]
df_top3 = df_all.sort_values(by = ['得票率'], ascending = False)[2:3]
rate1 = [100.0 - float(df_top1.iloc[0, 2])]
rate1.append(float(df_top1.iloc[0, 2]))
groups1 = ['非韓粉']
groups1.append(df_top1.iloc[0, 0])

rate2 = [100.0 - float(df_top2.iloc[0, 2])]
rate2.append(float(df_top2.iloc[0, 2]))
groups2 = ['非韓粉']
groups2.append(df_top2.iloc[0, 0])

rate3 = [100.0 - float(df_top3.iloc[0, 2])]
rate3.append(float(df_top3.iloc[0, 2]))
groups3 = ['非韓粉']
groups3.append(df_top3.iloc[0, 0])

fig = plt.gcf()
plt.subplot(1,3,1)
plt.pie(rate1, labels = groups1, autopct = '%0.01f%%', colors = ['r', 'b'])
plt.subplot(1,3,2)
plt.pie(rate2, labels = groups2, autopct = '%0.01f%%', colors = ['r', 'b'])
plt.subplot(1,3,3)
plt.pie(rate3, labels = groups3, autopct = '%0.01f%%', colors = ['r', 'b'])
fig.set_size_inches(20, 10)

plt.title('韓總信心指數最高區域 : 茂林',x = -0.7, y = -0.1, fontsize = 30)
```
