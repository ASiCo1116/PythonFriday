# 回歸一下國內生產毛額(GDP)跟出國人數的關聯
資料來源1：[中華民國統計資訊網] (http://statdb.dgbas.gov.tw/pxweb/Dialog/varval.asp?ma=NA8101A1A&ti=%B0%EA%A5%C1%A9%D2%B1o%B2%CE%ADp%B1%60%A5%CE%B8%EA%AE%C6-%A6%7E&path=../PXfile/NationalIncome/&lang=9&strList=L)

資料來源2：[交通部統計查詢網] (https://stat.motc.gov.tw/mocdb/stmain.jsp?sys=100&funid=a7101)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.font_manager import FontProperties
%matplotlib inline
```

## 先讀進出國人數資料

```python
df1 = pd.read_excel("__xls7241.xls", 2)
```

```python
df1
```

## 作為Y值所用

```python
y_train = df1.iloc[0:23, 1]#1986~2008
y_real = df1.iloc[23:, 1]
```

```python
y_train = np.array(y_train, dtype = np.int64)
y_real  = np.array(y_real, dtype = np.int64)
```

```python
y_real
```

## 再引入GDP資料

```python
df2 = pd.read_csv('NA8101A1Ac.csv', encoding = 'Big5')
```

```python
df2
```

## 當X資料用

```python
x_train = df2.iloc[2:25]
x_test = df2.iloc[25:35]
```

```python
x_train = np.array(x_train, dtype = np.int64)
x_test = np.array(x_test, dtype = np.int64)
```

# 預測一下吧！！

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
```

```python
regr.fit(x_train, y_train)
```

```python
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams.update({'font.size': 16})
```

```python
plt.scatter(x_train, y_train)
plt.title('人數v.sGDP(1986~2008)')
plt.xlabel('國內生產毛額, GDP(百萬元)')
plt.ylabel('國人出國人數(人次)')
# ax = plt.gca()
# for i, txt in enumerate(n):
#     ax.annotate(txt, (z[i], y[i]))
# plt.xlim(2500000, 1500000)
# plt.ylim(700000, 10000000)
```

```python
y_pred = regr.predict(x_test)
```

```python
plt.scatter(x_test, y_real)
plt.plot(x_test, y_pred, 'r')
plt.title('預測人數v.sGDP(2009~2018)')
plt.xlabel('國內生產毛額, GDP(百萬元)')
plt.ylabel('國人出國人數(人次)')
```

# 不難看出 在2009年之後 實際出國人數成長略呈指數成長
# 以簡單的線性回歸難以做到預測100%

```python
plt.scatter(np.append(x_train, x_test), np.append(y_train, y_real))
plt.title('實際人數v.sGDP(1986~2018)')
plt.xlabel('國內生產毛額, GDP(百萬元)')
plt.ylabel('國人出國人數(人次)')
```
