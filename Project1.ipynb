# 期中專題：陪猜(拳)機器人
### 有趣點之一：寫它的時候很有趣(而且不難)
### 有趣點之二：它可以陪邊緣人猜拳
### 有趣點之三：它還可以改成其他機器人(暫時想不到)
#### 目前擁有的功能：看得到機器人出什麼拳、計算連勝、勝率...待續

```python
import numpy as np
import tkinter as tk
import random 
```

```python
class Play:   
    def __init__(self, root):
        self.root = root
        root.title("RockPaperScissors")
        root.geometry('500x500')
    
        
        self.me = tk.StringVar()
        self.com = tk.StringVar()
        self.win = tk.StringVar()
        self.winNum = 0
        self.winStreak = 0
        self.game = 0
        self.winRate = tk.StringVar()
        
        self.computer = tk.Label(root, text = 'Computer :...', width = 30, height = 3)
        self.computer.grid(row = 0, column = 3)
        
        self.s1 = tk.Label(root, textvariable = self.me, width = 30, height = 3)
        self.s1.grid(row = 0, column = 0)
        
        self.s2 = tk.Label(root, textvariable = self.com, width = 30, height = 3)
        self.s2.grid(row = 1, column = 3)
        
        self.s3 = tk.Label(root, text = 'Win Streak: ', width = 30, height = 3)
        self.s3.grid(row = 5, column = 3)
        
        self.s4 = tk.Label(root, textvariable = self.win, width = 30, height = 3)
        self.s4.grid(row = 6, column = 3)
        
        self.s5 = tk.Label(root, text = 'Win Rate: ', width = 30, height = 3)
        self.s5.grid(row = 7, column = 3)
        
        self.s6 = tk.Label(root, textvariable = self.winRate, width = 30, height = 3)
        self.s6.grid(row = 8, column = 3)
        
        bS = tk.Button(self.root, text = 'Scissors', command = self.scissors)
        bP = tk.Button(self.root, text = 'Paper', command = self.paper)
        bR = tk.Button(self.root, text = 'Rock', command = self.rock)
        bS.grid(row = 2, column = 0, columnspan = 7, ipadx = 40, padx = 10)
        bP.grid(row = 3, column = 0, columnspan = 7, ipadx = 40, padx = 10)
        bR.grid(row = 4, column = 0, columnspan = 7, ipadx = 50, padx = 10)
        
    def scissors(self):
        self.game += 1
        computer = random.choice(['s', 'p', 'r'])
        if computer == 's':
            self.com.set('Scissors Out!')
            self.me.set('Tie!')
        elif computer == 'p':
            self.com.set('Paper Out!')
            self.me.set('You Win!')
            self.winNum += 1
            self.winStreak += 1
            self.win.set(str(self.winStreak))
            self.winRate.set(str(round(self.winNum/self.game, 3)))
        else:
            self.com.set('Rock Out!')
            self.me.set('You lose!')
            self.winStreak = 0
            self.win.set(str(self.winStreak))
            self.winRate.set(str(round(self.winNum/self.game, 3)))
    def paper(self):
        self.game += 1
        computer = random.choice(['s', 'p', 'r'])
        if computer == 'p':
            self.com.set('Paper Out!')
            self.me.set('Tie!')
        elif computer == 'r':
            self.com.set('Rock Out!')
            self.me.set('You Win!')
            self.winNum += 1
            self.winStreak += 1
            self.win.set(str(self.winStreak))
            self.winRate.set(str(round(self.winNum/self.game, 3)))
        else:
            self.com.set('Scissors Out!')
            self.me.set('You lose!')
            self.winStreak = 0
            self.win.set(str(self.winStreak))
            self.winRate.set(str(round(self.winNum/self.game, 3)))
    def rock(self):
        self.game += 1
        computer = random.choice(['s', 'p', 'r'])
        if computer == 'r':
            self.com.set('Rock Out!')
            self.me.set('Tie!')
        elif computer == 's':
            self.com.set('Scissors Out!')
            self.me.set('You Win!')
            self.winNum += 1
            self.winStreak += 1
            self.win.set(str(self.winStreak))
            self.winRate.set(str(round(self.winNum/self.game, 3)))
        else:
            self.com.set('Paper Out!')
            self.me.set('You lose!')
            self.winStreak = 0
            self.win.set(str(self.winStreak))
            self.winRate.set(str(round(self.winNum/self.game, 3)))
```

```python
window = tk.Tk()
gui = Play(window)
window.mainloop()
```
